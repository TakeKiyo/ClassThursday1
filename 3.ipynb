```python
import matplotlib.pyplot as plt
 
#図の大きさとフォントサイズを設定
plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
```

```python
N = [0.11]
P = [0.1]
alpha = 0.2
T = 10000
r=0.1
s=0.1
TimeList = []
```

```python
for t in range(T):
    TimeList.append(t)
    nt_1 = N[t]+alpha*N[t]*(r-P[t])
    pt_1 = P[t]+alpha*P[t]*(N[t]-s)
    N.append(nt_1)
    P.append(pt_1)
TimeList.append(T)
```

```python
# 1) 被食者の個体密度のグラフ
plt.plot(TimeList,N)
plt.xlabel("T")
plt.ylabel("N density")
plt.show()
```

```python
# 1) 捕食者の個体密度
plt.plot(TimeList,P)
plt.xlabel("T")
plt.ylabel("P density")
plt.show()
```

```python
# 2) 被食者の個体数と捕食者の個体数
# 最初の個体数を100000とする
num_n = [dens*100000 for dens in N]
num_p = [dens*100000 for dens in P]
plt.plot(num_n,num_p,'.')
plt.xlabel("N")
plt.ylabel("P ")
plt.show()
```
