```python
import matplotlib.pyplot as plt
 
#図の大きさとフォントサイズを設定
plt.rcParams["font.size"] = 20
plt.rcParams['figure.figsize'] = 12,10 
```

```python
NA = [0.1]
NB = [0.1]
P = [0.1]
alpha = 0.05
T = 10000
r_A=0.05
r_B=0.04
d_A = 0.8
d_B = 0.2
s=0.1
TimeList = []
```

```python
for t in range(T):
    TimeList.append(t)
    x = NA[t] + NB[t]
    q = d_A*NA[t] + d_B*NB[t]
    nAt_1 = NA[t]+alpha*NA[t]*(r_A-x-d_A*P[t])
    nBt_1 = NB[t]+alpha*NB[t]*(r_B-x-d_B*P[t])
    pt_1 = P[t]+alpha*P[t]*(q-s)
    NA.append(nAt_1)
    NB.append(nBt_1)
    P.append(pt_1)
TimeList.append(T)
```

```python
plt.plot(TimeList, NA)
plt.plot(TimeList, NB)
plt.plot(TimeList, P)
plt.xlabel("time")
plt.ylabel("population size")
plt.legend(["NA", "NB","P"])
plt.show()
```
